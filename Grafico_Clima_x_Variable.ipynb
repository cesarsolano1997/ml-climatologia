```python
import pandas as pd
```

```python
data_clima = pd.ExcelFile('D:/bibliografia de tesis/Datos/Datos Climaticos Viru Palto.xlsx')
df_clima = data_clima.parse('TABLA')                          
#df_clima
```

## Limpieza de datos

```python
df_clima.columns
```

```python
df_clima = df_clima[df_clima['Fecha'] >= '2019-01-01']
df_clima = df_clima[['Fecha','ET(mm).','Tº MIN/DIA (°C)','Tº MAX/DIA (°C)','Rad. Solar Prom.','HUMEDAD PROM.']]
#df_clima
```

```python
df_clima = df_clima.rename(columns = {'ET(mm).' : 'ET', 
                                      'Tº MIN/DIA (°C)' : 'T_MIN',
                                     'Tº MAX/DIA (°C)': 'T_MAX',
                                     'Rad. Solar Prom.' : 'RS_PROM',
                                     'HUMEDAD PROM.': 'HM_PROM'})
df_clima
```

```python
# Importando matplotlib
import matplotlib.pyplot as plt
```

```python
for vr_cl in df_clima.columns:
    
    if vr_cl != 'Fecha':
        x = df_clima['Fecha']
        y = df_clima[vr_cl]
        

        plt.figure(figsize=(20,6))
        plt.title('Gráfica de ' + str(vr_cl) + ' año 2019')
        plt.xlabel('Fecha')
        plt.ylabel('Evapotranspiración')
        plt.plot(x,y)
        plt.savefig(str(vr_cl) + '_2019.png')
        plt.show()
```
