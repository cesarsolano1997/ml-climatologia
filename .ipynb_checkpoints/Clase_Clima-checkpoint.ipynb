```python
#Importando librerias
import pandas as pd       
import matplotlib.pyplot as plt
```

```python
class Weather:       
    
    def ImportExcel(path, leaf):
        data_clima = pd.ExcelFile(path)
        global dfClima 
        dfClima = data_clima.parse(leaf)  
    def ClearData(fecha):
        global dfClima 
        flClima = dfClima[dfClima['Fecha'] >= fecha]
        flClima = dfClima[['Fecha','ET(mm).','Tº MIN/DIA (°C)','Tº MAX/DIA (°C)','Rad. Solar Prom.','HUMEDAD PROM.']]   
        flClima = flClima.rename(columns = {'ET(mm).' : 'ET', 
                                      'Tº MIN/DIA (°C)' : 'T_MIN',
                                     'Tº MAX/DIA (°C)': 'T_MAX',
                                     'Rad. Solar Prom.' : 'RS_PROM',
                                     'HUMEDAD PROM.': 'HM_PROM'})
        dfClima = flClima
    def Chart():
        global dfClima 
        for vr_cl in dfClima.columns:
    
            if vr_cl != 'Fecha':
                x = dfClima['Fecha']
                y = dfClima[vr_cl]


                plt.figure(figsize=(20,6))
                plt.title('Gráfica de ' + str(vr_cl) + ' año 2019')
                plt.xlabel('Fecha')
                plt.ylabel('Evapotranspiración')
                plt.plot(x,y)
                plt.savefig(str(vr_cl) + '_2019.png')
                plt.show()
    def Imprimiendo():
        print(dfClima)
```

```python
#Weather.init()
```

```python
Weather.ImportExcel('D:/bibliografia de tesis/Datos/Datos Climaticos Viru Palto.xlsx','TABLA')
```

```python
Weather.ClearData('2019-01-01')
```

```python
Weather.Chart()
```

```python
Weather.Imprimiendo()
```
