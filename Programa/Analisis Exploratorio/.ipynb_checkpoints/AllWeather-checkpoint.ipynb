# Limpieza de datos del clima

```python
import pandas as pd
import numpy as np
```

```python
exClima = pd.ExcelFile('D:/bibliografia de tesis/Datos/Datos Climaticos Viru Palto.xlsx')
```

```python
dfClima = exClima.parse('TABLA')
```

```python
dfClima.head()
```

```python
dfClima = dfClima[(dfClima['Año'] == 2019)]
```

```python
dfClima = dfClima[['Fecha','ET(mm).','Tº MIN/DIA (°C)','Tº MAX/DIA (°C)','Rad. Solar Prom.','HUMEDAD PROM.', 'T° Prom/Día (°C)']]
dfClima = dfClima.rename(columns = {'ET(mm).' : 'ET', 
                                      'Tº MIN/DIA (°C)' : 'T_MIN',
                                     'Tº MAX/DIA (°C)': 'T_MAX',
                                     'Rad. Solar Prom.' : 'RS_PROM',
                                     'HUMEDAD PROM.': 'HM_PROM',
                                      'T° Prom/Día (°C)': 'TMP_PROM'})
```

```python
dfClima.reset_index(drop = True, inplace = True)
dfClima.sort_values(by=['Fecha'])
dfClima.head()
```

```python
#dfClima.to_excel('clearclima.xlsx')
```

## Graficando y guardando información en carpetas

```python
objNombClima = { 'ET': 'Evapotranspiración', 'T_MIN': 'Temperatura mínima', 'T_MAX' : 'Temperatura máxima', 'RS_PROM' : 'Radiación solar promedio', 'HM_PROM' : 'Humedad Promedio' , 'TMP_PROM': 'Temperatura promedio' }
```

```python
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
arrVarClima = dfClima.columns
for varClima in arrVarClima:  
    if varClima != 'Fecha':                
        
        x = dfClima['Fecha']
        y = dfClima[varClima]   

        plt.rcParams["figure.figsize"] = (20,5)
        
        plt.title(objNombClima[varClima], fontdict={ 'fontsize' : 30})
        plt.xlabel('Fecha')
        plt.ylabel(varClima)                    
        plt.plot(x,y, label = varClima)
        plt.show()
```
