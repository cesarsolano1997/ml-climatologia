## Limpieza de datos

```python
#Filtrar por fecha
fl_df_Clima =  dfClima[dfClima['Fecha'] >= '2019-01-01']
fl_df_Clima.columns

```

```python
df_Clima_ET = fl_df_Clima[['Fecha','ET(mm).']]
#df_Clima_ET
```

```python
# Importando matplotlib
import matplotlib.pyplot as plt
```

```python
x = df_Clima_ET["Fecha"]
y = df_Clima_ET["ET(mm)."]
```

```python
# Configuración del gráfico
plt.figure(figsize=(20,6))
plt.title('Grafica de Evapotranspiración año 2019')
plt.xlabel('Fecha')
plt.ylabel('Evapotranspiración')


# Graficando
plt.plot(x,y)

plt.savefig('ET_2019.png')
plt.show()
```
